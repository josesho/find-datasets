```python
weightgain = get_csv(820)
weightgain.to_csv('weightgain.csv')

weightgain.columns = ['index', 'source', 'type', 'weight_gain_grams']
# Make new index.
index_fragment = [i for i in range(0,10)]
index_all = index_fragment * 4
weightgain.loc[:, 'index'] = index_all
# Concatenate source and type cols.
weightgain['diet'] = weightgain.source.str.cat(weightgain.type, sep=' ')
# Pivot.
weightgain_wide = weightgain.pivot('index','diet','weight_gain_grams')
# Re-arrange columns.
weightgain_wide = weightgain_wide[['Beef Low', 'Beef High', 'Cereal Low', 'Cereal High']]
```

```python
weightgain_wide
```

```python
f, c = bsc.contrastplot(weightgain_wide, 
                 paired=True,
                 float_contrast=False,
                 swarm_label='weight gain (g)',
                 contrast_label='delta paired\nweight gain (g)',
                 idx=(('Beef Low','Beef High'),
                      ('Cereal Low','Cereal High')) )
```
